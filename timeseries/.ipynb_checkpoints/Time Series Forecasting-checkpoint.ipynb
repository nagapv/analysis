### ADF test for stationarity.
> 
Augmented DickeyFuller test

```python
def _check_stationarity(data):
    from statsmodels.tsa.stattools import adfuller

    result = adfuller( data )
    
    print('ADF Statistic: {0}; p-value: {1}\n'.format(result[0], result[1]))
    if result[1] > 0.05:
        print("Failed to reject null hypothesis! Data has a unit root and is non-stationary\n")
    else:
        print("Rejected the null hypothesis! The data is stationary.\n")
```

```python
# Evaluate whole dataset
_check_stationarity( df['PM2.5'].values )
```

### Check for seasonality in the data.
> Long run.

```python
data = df.set_index("Timestamp")[['PM2.5']]
result = seasonal_decompose(data, model='additive', period=(24*30*12), extrapolate_trend='freq')
plt.figure(figsize=(6, 4))
plt.plot(result.seasonal)
```

> Daily.

```python
fig, ax = plt.subplots(figsize=(6, 4))

g = df.groupby(['Year','Month','Day','Hour'])['PM2.5'].mean().reset_index().tail(1000)
for y in g.Year.unique():
    for m in g.Month.unique():
        for d in g.Day.unique():
            _tf = g[ (g.Year==y) & (g.Month==m) & (g.Day==d) ]
            if _tf.empty:
                pass
            else:
                ax.plot(_tf.Hour.values, _tf['PM2.5'].values)
ax.set_title("Seasonality by Hour of the Day")
```

### Accounting for missing values.

```python
#
_tf = df[['Timestamp','PM2.5']]
# _tf.set_index('Timestamp').plot()
```

```python
dateset = []
dmin, dmax = _tf.Timestamp.min(), _tf.Timestamp.max()
daydelta = int( (dmax - dmin).days )

for _date in (dmin + timedelta(n) for n in range(daydelta)):
    for i in range(24):
        date = _date + timedelta(hours=i)
        dateset.append(date)
```

```python
_gdf = pd.DataFrame(dateset, columns=['Timestamp'])
_gdf = _gdf.merge(_tf, on='Timestamp', how='left')
_gdf.info()
```

```python
_missing_idx = _gdf[_gdf['PM2.5'].isnull()].index
len(_missing_idx), _missing_idx
```

```python
_gdf.sample()
```

```python
# Calculating moving averages; forward filling last value to avoid blanks.
gdf = _gdf.copy()
_mvg = gdf['PM2.5'].ffill().rolling(window=3*24, min_periods=0).mean()
```

```python
# Filling missing values with values generated above.
gdf['PM2.5'] = gdf['PM2.5'].mask(cond=gdf['PM2.5'].isnull(), other=_mvg, axis=0)
```

```python
gdf.info()
```

```python
# Evaluate imputed of the dataset
_check_stationarity( gdf['PM2.5'].values )
```

```python
gdf.set_index('Timestamp').plot()
```

```python
# # This code is  for KFold splits
# tscv = TimeSeriesSplit(n_splits=2, test_size=50)

# X, y = gdf['Timestamp'], gdf['PM2.5']

# for fold, (train_index, test_index) in enumerate(tscv.split(gdf)):
#     print("Fold: {}".format(fold))
#     print("train indices:", train_index, "\n", "test indices:", test_index)
#     print("\n")
#     X_train, X_test = X[train_index], X[test_index]
#     y_train, y_test = y[train_index], y[test_index]
```

```python
X, y = gdf['Timestamp'], gdf['PM2.5']

X_test, y_test = X.tail(50), y.tail(50)
X_train, y_train = X[X.index.isin(X_test.index)==False], y[y.index.isin(y_test.index)==False]

print(X_train.index, '\n')
print(X_test.index)
```

> [ACF and PACF](https://spureconomics.com/interpreting-acf-and-pacf-plots/)

```python
fig = plt.figure(figsize=(10, 7))

ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(gdf['PM2.5'].values.squeeze(), lags=50, ax=ax1)

ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(gdf.set_index('Timestamp'), lags=50, ax=ax2)
```

### ARIMA Model and ARMA Model

p: The number of lag observations included in the model, also called the lag order (deals with window of Xt)    
      
d: The number of times that the raw observations are differenced, also called the degree of differencing (deals with order of differencing of Xt)    
      
q: The size of the moving average window, also called the order of moving average (deals with residuals)     
       
Description from - https://phdinds-aim.github.io/time_series_handbook/01_AutoRegressiveIntegratedMovingAverage/01_AutoRegressiveIntegratedMovingAverage.html

> ARIMA (primary model)

```python
# The order=(p,d,q) order of the model for the autoregressive, differences, and moving average components.
model1 = ARIMA(y_train, order=(14, 1, 24))
model_fit1 = model1.fit()
print(model_fit1.summary())
```

> ARMA (without differencing for comparision)

```python
model2 = ARIMA(y_train, order=(11, 0, 10))
model_fit2 = model2.fit()
# print(model_fit2.summary())
```

```python
actual_values = _gdf[_gdf.index.isin([i for i in range(40030, 40080)])]['PM2.5'].values
len(actual_values), len(X_test)
```

```python
# Forecast and compare
fig, ax = plt.subplots(figsize=(10,4))

forecast = model_fit1.predict(40030, 40079)

ax.plot(X_test, actual_values, label="Actual value", color='#33a02c')
ax.plot(X_test, forecast, label="Predicted [ARIMA]", color='#ff7f00')

ax.set_title("Air Quality Prediction, India")
ax.legend()
ax.set_xlabel("Month-Date, Hour")
ax.set_ylabel("PM 2.5")

plt.savefig("air_quality_prediction_india.png")
```

```python
# Forecast and compare
fig, ax = plt.subplots(figsize=(10,4))
forecast1 = model_fit1.predict(40030, 40079)
forecast2 = model_fit2.predict(40030, 40079)
ax.plot(X_test, actual_values, label="Actual value", color='#33a02c')
ax.plot(X_test, forecast1, label="Predicted [ARIMA]", color='#ff7f00')
ax.plot(X_test, forecast2, label="Predicted [ARMA]", color='#a6cee3', alpha=0.5)
ax.legend()
plt.title("for comparision")
```
